```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = m = np.arange(0,11,1)
a = 1
x1 = n*a
y1 = m*a
x2 = (n+0.5)*a
y2 = (m+0.5)*a
L = a /2
X1,Y1 = np.meshgrid(x1,y1)
X2,Y2 = np.meshgrid(x2,y2)
theta1 = 0  # rod orientation (e.g., 45 degrees)
theta2 = np.pi/2  # rod orientation (e.g., 45 degrees)

# Rod vector components (half-length in each direction)
dx1 = (L / 2) * np.cos(theta1)
dy1 = (L / 2) * np.sin(theta1)
dx2 = (L / 2) * np.cos(theta2)
dy2 = (L / 2) * np.sin(theta2)
```

```python
n_rows = 10                 # number of rows
n_cols = 10                 # number of columns
dx = a * np.sqrt(3)/np.sqrt(2)         # horizontal spacing
dy = a * 1.5/np.sqrt(2)                # vertical spacing

# Generate coordinates
x_coords = []
y_coords = []

for row in range(n_rows):
    for col in range(n_cols):
        x = col * dx + (dx / 2 if row % 2 else 0)  # offset every other row
        y = row * dy
        x_coords.append(x)
        y_coords.append(y)

fig, (ax1,ax2) = plt.subplots(1,2,figsize = (10,5))

for x0, y0 in zip(x_coords, y_coords):
    ax1.plot([x0 - dx1, x0 + dx1], [y0 - dy1, y0 + dy1], linewidth=2,color = 'orange')
for x0, y0 in zip(x_coords, y_coords):
    ax1.plot([x0 - dx2+0.5*a, x0 + dx2+0.5*a], [y0 - dy2+0.5*a, y0 + dy2+0.5*a], linewidth=2,color = 'blue')

for x0, y0 in zip(X1.ravel(), Y1.ravel()):
    ax2.plot([x0 - dx1, x0 + dx1], [y0 - dy1, y0 + dy1], linewidth=2,color = 'orange')

for x0, y0 in zip(X2.ravel(), Y2.ravel()):
    ax2.plot([x0 - dx2, x0 + dx2], [y0 - dy2, y0 + dy2], linewidth=2,color = 'blue')
ax1.set_aspect('equal')
ax2.set_aspect('equal')

ax2.plot([],color = 'blue',label = 'Top')
ax2.plot([],color = 'orange',label = 'Bottom')
ax2.legend()
plt.show()
```

```python
a1 = np.array([2, 0])
a2 = np.array([1, np.sqrt(3)])

# Two-atom basis (as fractions of lattice vectors)
x0_hc = np.array([[0, 0], [0, 2/np.sqrt(3)]])  # These will be placed relative to Bravais lattice

# Grid size
Nx, Ny = 6, 6  # Number of unit cells in x and y

# Generate full honeycomb lattice
points = []
for i in range(Nx):
    for j in range(Ny):
        R = i * a1 + j * a2
        for atom in x0_hc:
            pos1 = R + atom
            pos2 = R + atom
            points.append(pos1)
            points.append(pos2)

points = np.array(points)
plt.plot(points[:, 0], points[:, 1], 'ko', markersize=4)
plt.axis('equal')
plt.show()
```
